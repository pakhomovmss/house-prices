```python
# look at some scatter

def get_scatter(df, var, clr='blue'):
    plt.figure(figsize=(10,7))
    plt.scatter(x=df[var], y=df["SalePrice"], marker='o', color=clr)
    plt.xlabel(var)
    plt.ylabel("SalePrice")
    plt.show()
```

```python
get_scatter(df_train, 'GrLivArea')
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x=df_train["LotArea"], y=np.log(df_train["SalePrice"]), marker='o', color='purple')
plt.xlabel("LotArea")
plt.ylabel("SalePrice")
plt.show()
```

```python
get_scatter(df_train, 'TotalBsmtSF', clr='green')
```

```python
plt.figure(figsize=(18, 9))
sns.boxplot(x='YearBuilt', y='SalePrice', data=df_train)
plt.xticks(rotation=90)
plt.show()
```

```python
# find out some correlation between features

corrmat = df_train.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corrmat, vmax=0.8, cbar=True, square=True).set_title("Correlation Matrix")
```

```python
max_feat = 15

cols = corrmat.nlargest(max_feat, "SalePrice")["SalePrice"].index
cm = np.corrcoef(df_train[cols].values.T)
sns.set(font_scale=1.25)

plt.figure(figsize=(12,12))
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)

bottom, top = hm.get_ylim()
hm.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```
