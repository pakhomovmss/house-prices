```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, chi2

import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv("E:\\Data Science\\Datasets\\HousePrices\\train_clear.csv")
```

```python
train
```

```python
del [train["Id"], train["OverallQual"], train["GrLivArea"], train["GarageCars"], train["TotalBsmtSF"], train["FullBath"], train["YearBuilt"]]
```

```python
train
```

```python
#train.drop(train[train["GrLivArea"] > 4500].index.values, inplace=True)
#train.drop(train[train["TotalBsmtSF"] > 6000].index.values, inplace=True)
```

```python
def max_len(arr):
    lengths = []
    for i in arr:
        lengths.append(len(i))
    return max(lengths)
```

```python
count_factor = 0
count_num = 0

factor_cols = []

for col in train.columns:
    
    print(col + (max_len(train.columns) - len(col) + 4) * " " + str(train[col].dtype))
    
    if train[col].dtype == "object":
        factor_cols.append(col)
        count_factor += 1
    
    else:
        count_num += 1

print("\nNumeric: " + str(count_num), "    Factor: " + str(count_factor))
```

```python
#train["MasVnrArea"] = train["MasVnrArea"].astype("float")
#set(train["MasVnrArea"])
train["MasVnrArea"].replace({'No' : 0}, inplace=True)
```

```python
train["MasVnrArea"] = train["MasVnrArea"].astype("float")
```

```python
#train["GarageYrBlt"] = train["GarageYrBlt"].astype("float")
#train[train["GarageYrBlt"] == 'No']
set(train["GarageYrBlt"])
train['NoGarage'] = 0
train.loc[train["GarageYrBlt"] == 'No', ["NoGarage"]] = 1
```

```python
del train['GarageYrBlt']
```

```python
train
```

```python
data = pd.get_dummies(train)
```

```python
data
```

```python
label = data.pop("SalePrice")
```

```python
columns = np.array(data.columns)
```

```python
columns
```

```python
label = np.array(label)
```

```python
data = np.array(data)
```

```python
data.shape
```

```python
selector = SelectKBest(score_func=chi2, k=20)
```

```python
score = selector.fit(data, label)
```

```python
score.scores_.shape
```

```python
feature_importance = dict(zip(score.scores_, columns))
feature_importance
```

```python
importances = sorted(feature_importance.keys())[::-1][:17]
features = []

for imp in importances:
    features.append(feature_importance[imp])

features
```

```python
set(train.MiscVal)
```

```python
df = pd.get_dummies(train)
data_to_model = df[features].copy()
```

```python
test = pd.read_csv("E:\\Data Science\\Datasets\\HousePrices\\test.csv")
```

```python
t = pd.get_dummies(test)
```

```python
del t['Id']
```

```python
test_to_model = t[features].copy()
```

```python
test_to_model.isna().sum()
```

```python
test_to_model['MasVnrArea'].fillna(test_to_model['MasVnrArea'].mode().max(), inplace=True)
test_to_model['TotalBsmtSF'].fillna(test_to_model['TotalBsmtSF'].mode().max(), inplace=True)
test_to_model['GarageArea'].fillna(test_to_model['GarageArea'].mode().max(), inplace=True)
```

```python
data_to_model
```

```python
def get_scatter(df, var, clr='blue'):
    plt.figure(figsize=(10,7))
    plt.scatter(x=df[var], y=df["SalePrice"], marker='o', color=clr)
    plt.xlabel(var)
    plt.ylabel("SalePrice")
    plt.show()
```

```python
get_scatter(data_to_model, "GrLivArea")
```

```python
data_to_model['SalePrice'] = np.log(data_to_model['SalePrice'])
```

```python
def normalize(df, col):
    df[col] = (df[col].values - df[col].mean()) / df[col].std()
```

```python
normalize(data_to_model, 'TotalBsmtSF')
```

```python
data_to_model['GrLivArea'] = np.log(data_to_model['GrLivArea'])
```
