```python
def separate_by_LotFrontage_fillna(df):
    
    container = df[(df["Neighborhood"] == 'Blmngtn') & (df["LotFrontage"].isna())]
    mode = df[df["Neighborhood"] == 'Blmngtn'].LotFrontage.mode()[0]
    container["LotFrontage"].fillna(mode, inplace=True)
    
    for district in set(df["Neighborhood"]):
        if district != 'Blmngtn':
            mode = df[df["Neighborhood"] == district].LotFrontage.mode()[0]
            container = pd.concat([container, df[(df["Neighborhood"] == district) & (df["LotFrontage"].isna())]], axis=0)
            container["LotFrontage"].fillna(mode, inplace=True)
            
    return container
```

```python
container = separate_by_LotFrontage_fillna(df_train)

ids_to_del = np.array(container.Id) - 1

df_train.drop(ids_to_del, inplace=True)

df_train = pd.concat([df_train, container], axis=0)
df_train.sort_index(inplace=True)
```

```python
def get_similar_columns_list(df, keyword):
    return [i for i in (df_train.columns.str.contains(keyword) * np.array(df_train.columns)).tolist() if i != '']

def min_rows(df, cols):
    
    minimal = df[df[cols[0]].isna()].shape[0]
    min_ind = 0
    
    for i, col in enumerate(cols):
        if df[df[col].isna()].shape[0] < minimal:
            minimal = df[df[col].isna()].shape[0]
            min_ind = i
    
    return min_ind

def is_columns_identical(df, col_list, indices):
    cols = []
    conditions = df[col_list[0]].isna()
    
    for col in indices:
        cols.append(col_list[col])
        if col > 0:
            conditions = conditions & df[col_list[col]].isna()
    
    return [df[conditions].shape[0] == df[df[col_list[min_rows(df, cols)]].isna()].shape[0], cols]    
```

```python
# !!! !!! !!!

cmp_res = is_columns_identical(df_train, get_similar_columns_list(df_train, "Bsmt"), indices=[0,1,2,3,5])

# ИСПРАВИТЬ!!! Заполняет все колонки полностью, хотя в BsmtExposure и BsmtFinType2 должно остаться еще по одному Nan !!!
# !!!
if cmp_res[0]:
    for column in cmp_res[1]:
        df_train[column].fillna("No", inplace=True)
        
show_missing_columns(df_train, percent=True)
```
